```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Univariate Analysis (KDE Plots)
numerical_columns = MD_agric_df.select_dtypes('number').drop(columns=["Field_ID"])
fig, axes = plt.subplots(4, 4, figsize=(15, 15))
axes = axes.flatten()

for i, column in enumerate(numerical_columns.columns):
    sns.kdeplot(data=MD_agric_df, x=column, ax=axes[i], fill=True)
    axes[i].axvline(MD_agric_df[column].mean(), color='red', linestyle='dashed')
    axes[i].set_title(f'Dist: {column}')

plt.tight_layout()
plt.show()

# 2. Categorical Analysis (Violin Plot)
plt.figure(figsize=(12, 7))
sns.violinplot(data=MD_agric_df, x='Crop_type', y='Rainfall')
plt.title('Rainfall Distribution by Crop Type')
plt.xticks(rotation=45)
plt.show()
```
